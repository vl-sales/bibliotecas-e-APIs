### Preencher os dados faltantes
- Prós - Não perde dados
- Contra - São usados dados artificiais

Algumas formas:

- Preencher com a Moda: Elemento mais comum.
- Preencher com a média;
- Preencher com a mediana.


Considerações especiais:

Quando há uma relação temporal é legal usar os seguintes metodos:
- foward fill - Propagar o último valor presente ao longo do trecho faltante
- Interpolação - Fazer uma de crescimento linear
- bfill (backward fill) - quando o período faltante é o inicial

```python
df['Age'].fillna(df['Age'].mean())
```
