## Múltiplas Curvas

````python
x = np.arange(0, 10, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)
````

````python
plt.plot(x, y1, x, y2)
````

````python
plt.plot(x, y1, label='Sen(x)')
plt.plot(x, y2, label='Cos(x)')
plt.legend()
````

````python
# Mais de um axes
plt.subplot(2, 1, 1) #Cria região para plotar a primeira curva
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.plot(x, y2)

plt.tight_layout() #Organiza os axes
````

````python
y3 = np.sin(x) + np.cos(3*x)
y4 = np.cos(x) + np.sin(5*x)
````

````python
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.plot(x, y2)

plt.subplot(2, 1, 2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.tight_layout()
````

````python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Axes 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Axes 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Axes 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Axes 4')

plt.tight_layout()
````

````python
#Forma 2
figure, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x, y1)
axes[0, 0].set_title('Axes 1')
axes[0, 0].set_xlabel('X')

axes[1,1].plot(x, y4)
axes[1,1].set_title('Axes 4')

plt.tight_layout()
````

````python
x = np.arange(1, 5, 0.01)
y1 = (np.sin(x) / x)
y2 = 2 ** x / 20
````

````python
plt.style.use('seaborn') # ggplot, dark_background, seaborn, etc
plt.plot(x, y1, label='X1')
plt.plot(x, y2, label='X2')
plt.title('Primeiro exercício', fontsize=20, fontfamily='serif', fontweight='bold')
plt.legend(fontsize=15, loc='upper left')
````

````python
plt.subplot(2, 1, 1)
plt.plot(x, y1, label='X1')
plt.legend()
plt.title('Primeiro Ex.')

plt.subplot(2, 1, 2)
plt.plot(x, y2, label='X2')
plt.legend()
plt.title('Primeiro Ex. pt.2')
````

````python
figure, axes = plt.subplots(2, 1)

axes[0].plot(x, y1)
axes[0].set_title('X1')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(x, y2)
axes[1].set_title('X2')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
````

### API Orientada a Objetos

````python
plt.style.use('ggplot')
figure = plt.figure(figsize=(6, 6)) #Onde fica abrigado os axes
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height

plt.style.use('dark_background')
axes2 = figure.add_axes([0.65, 0.65, 0.2, 0.2])
plt.xticks(color='black')
plt.yticks(color='black')

axes.plot(x, y1)
axes2.plot(x, y2, color='red')

axes.axis([1, 5, -1, 3])
circle = plt.Circle((3, 0.2), 0.3)
axes.add_artist(circle)
#Circulo adicionado apenas para fins didáticos
````
